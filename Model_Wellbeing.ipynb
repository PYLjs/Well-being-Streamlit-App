# Data preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Wellbeing.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['DAILY_STRESS'].value_counts()
```

```python
df.loc[df['DAILY_STRESS'] == '1/1/00']
```

```python
df['DAILY_STRESS'] = df['DAILY_STRESS'].replace('1/1/00',int(3))
df['DAILY_STRESS'] = df['DAILY_STRESS'].astype('int64')
```

```python
df['DAILY_STRESS'].value_counts()
```

### I want the work life balance score to be between 0 and 100 

```python
df['WORK_LIFE_BALANCE_SCORE'] = (df['WORK_LIFE_BALANCE_SCORE']-df['WORK_LIFE_BALANCE_SCORE'].min())/(df['WORK_LIFE_BALANCE_SCORE'].max()-df['WORK_LIFE_BALANCE_SCORE'].min())*100 
```

```python
df['WORK_LIFE_BALANCE_SCORE'].describe()
```

```python
df = df.drop('Timestamp',axis=1)
```

```python
df.to_csv(r'C:\Users\33615\OneDrive\Documents\datasets\wellbeing.csv')
```

```python
df['AGE'].value_counts()
```

```python
df['AGE'] = df['AGE'].replace('Less than 20',int(0))
df['AGE'] = df['AGE'].replace('21 to 35',int(1))
df['AGE'] = df['AGE'].replace('36 to 50',int(2))
df['AGE'] = df['AGE'].replace('51 or more',int(3))

df['GENDER'] = df['GENDER'].replace('Male',int(0))
df['GENDER'] = df['GENDER'].replace('Female',int(1))
```

```python
df.head()
```

```python
df.describe()
```

### Now that we have cleaned and export our dataset as CSV file, we can build the model.

```python
from sklearn.model_selection import train_test_split, GridSearchCV

y = df['WORK_LIFE_BALANCE_SCORE']
X = df.drop('WORK_LIFE_BALANCE_SCORE',axis=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
param_grid = {'n_estimators':[10, 50, 100, 200],
              'max_depth':[3, 5, 7, 10, 15, 20, None]}

grid = GridSearchCV(rfr,param_grid,cv=5)
grid.fit(X_train,y_train)
params = grid.best_params_
```

```python
print(f"The best parameters are : {params}")
```

```python
model = RandomForestRegressor(max_depth=15,n_estimators=200)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
